```python
import urllib.request, urllib.error, urllib.parse
```

```python
year = "2019"
month = "01"
day = "01"
hour = "00"

url = f'https://s3-us-west-1.amazonaws.com//files.airnowtech.org/airnow/{year}/{year}{month}{day}/HourlyData_{year}{month}{day}{hour}.dat'

response = urllib.request.urlopen(url)
content = response.read()

f = open(f'../data/raw/HourlyData_{year}{month}{day}{hour}.dat', 'wb')
f.write(content)
f.close
```
