```python
import pandas as pd
```

```python
df = pd.read_csv('../data/preprocessed/hourly_AQ_data.csv')
df.info()
```

```python
df = df.drop(df.columns[[0, 1, 4, 6, 10]], axis=1)
```

```python
df_sum = df.groupby(['date', 'parameter', 'units']).agg({'value':['min', 'mean', 'max', 'std']})
df_sum.columns = ["_".join(x) for x in df_sum.columns.ravel()]
```

```python
df_sum.to_csv('../data/preprocessed/daily_AQ_data.csv')
```
