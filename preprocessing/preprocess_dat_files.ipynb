```python
import os
import glob
import pandas as pd
from datetime import datetime, timedelta
```

```python
path = r'../data/raw/hourly_data/'
```

```python
columns = ['date', 'time', 'aqsid', 'site', 'gmtoffset', 'parameter', 'units', 'value', 'source']
```

```python
files2018 = glob.glob(os.path.join(path, "HourlyData_2018*.dat"))
files20190 = glob.glob(os.path.join(path, "HourlyData_20190*.dat"))
files20191 = glob.glob(os.path.join(path, "HourlyData_20191*.dat"))
```

```python
df2018 = pd.concat((pd.read_csv(f, sep='|', header=None, names=columns) for f in files2018), ignore_index=True)
```

```python
df2018.info()
```

```python
df2018 = df2018[df2018.source.str.startswith('New Y')]
```

```python
df2018.info()
```

```python
df2018.site.unique()
```

```python
boroughs = ['Bronx - IS52', 'Pfizer Lab', 'Bklyn - PS 314','Manhattan/IS143',
       'Division Street', 'CCNY', 'Queens', 'Queens Near-road', 'Fresh Kills',
       'Bklyn - PS274', 'IS52', 'PS 314', 'Brooklyn - PS274', 'IS143', 'Bronx - IS74', 'FKILL']
```

```python
df2018 = df2018[df2018.site.isin(boroughs)]
df2018.info()
```

```python
df2018.to_csv('../data/preprocessed/data_2018.csv')
```

```python
del df2018
```
